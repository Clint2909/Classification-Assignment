```python
import pandas as pd
```

```python
dataset=pd.read_csv('CKD.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset["classification_yes"].value_counts()
```

```python
dataset.columns
```

```python
independent=dataset[[ 'age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dependent=dataset[['classification_yes']]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
independent=sc.fit_transform(independent)
```

```python
from sklearn.model_selection  import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

param_grid={'criterion':['gini', 'entropy','log_loss'],'splitter':['best','random'],'max_features':['int','sqrt','log2']}

grid=GridSearchCV(DecisionTreeClassifier(),param_grid,refit=True,cv=5,verbose=3,n_jobs=-1,scoring='f1_weighted')
grid.fit(independent,dependent)
```

```python
re=grid.cv_results_

grid_predictions = grid.predict(independent) 

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(dependent, grid_predictions)

from sklearn.metrics import classification_report
clf_report = classification_report(dependent,grid_predictions)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(dependent,grid_predictions,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

```python
print("The confusion Matrix:\n",cm)
```

```python
print("The report:\n",clf_report)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(dependent,grid.predict_proba(independent)[:,1])
```
